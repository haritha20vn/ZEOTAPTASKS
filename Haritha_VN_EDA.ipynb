###Transactions
* TransactionID: Unique identifier for each transaction.
* CustomerID: ID of the customer who made the transaction.
* ProductID: ID of the product sold.
* TransactionDate: Date of the transaction.

* Quantity: Quantity of the product purchased.
* TotalValue: Total value of the transaction.
* Price: Price of the product sold.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "/content/drive/MyDrive/ZEOTAP/Transactions.csv"  # Replace with your file path
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("Dataset Overview:")
print(data.head())

# Dataset info
print("\nDataset Info:")
print(data.info())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Check for duplicate transactions
print("\nDuplicate Transactions:")
print(data.duplicated().sum())

# Summary statistics for numeric columns
print("\nSummary Statistics:")
print(data.describe())

# Convert TransactionDate to datetime
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'])

# Extract Year, Month, and Day for analysis
data['Year'] = data['TransactionDate'].dt.year
data['Month'] = data['TransactionDate'].dt.month
data['Day'] = data['TransactionDate'].dt.day

# Unique customers, products, and transactions
print("\nUnique Counts:")
print(f"Unique Transactions: {data['TransactionID'].nunique()}")
print(f"Unique Customers: {data['CustomerID'].nunique()}")
print(f"Unique Products: {data['ProductID'].nunique()}")

# Monthly transaction trends
monthly_trends = data.groupby(['Year', 'Month']).size().reset_index(name='TransactionCount')
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_trends, x='Month', y='TransactionCount', hue='Year', marker='o')
plt.title('Monthly Transaction Trends')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.tight_layout()
plt.show()

# Quantity vs. TotalValue
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Quantity', y='TotalValue', alpha=0.7, color='blue')
plt.title('Quantity vs. Total Value')
plt.xlabel('Quantity')
plt.ylabel('Total Value')
plt.tight_layout()
plt.show()

# TotalValue distribution
plt.figure(figsize=(8, 5))
sns.histplot(data['TotalValue'], kde=True, bins=30, color='green')
plt.title('Total Value Distribution')
plt.xlabel('Total Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Price vs. Quantity boxplot by year
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Year', y='Price', palette='cool')
plt.title('Price Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.tight_layout()
plt.show()

# Analyze outliers in TotalValue
print("\nOutliers in TotalValue:")
q1 = data['TotalValue'].quantile(0.25)
q3 = data['TotalValue'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = data[(data['TotalValue'] < lower_bound) | (data['TotalValue'] > upper_bound)]
print(outliers)

# Customer analysis: Top customers by TotalValue
top_customers = data.groupby('CustomerID')['TotalValue'].sum().nlargest(10).reset_index()
print("\nTop Customers by Total Value:")
print(top_customers)

# Visualize top customers
plt.figure(figsize=(8, 5))
sns.barplot(data=top_customers, x='CustomerID', y='TotalValue', palette='viridis')
plt.title('Top 10 Customers by Total Value')
plt.xlabel('CustomerID')
plt.ylabel('Total Value')
plt.tight_layout()
plt.show()

# Summary of observations
print("\nEDA Summary:")
print(f"Total Transactions: {data.shape[0]}")
print(f"Unique Customers: {data['CustomerID'].nunique()}")
print(f"Unique Products: {data['ProductID'].nunique()}")
print(f"Transaction Period: {data['TransactionDate'].min()} to {data['TransactionDate'].max()}")
```
